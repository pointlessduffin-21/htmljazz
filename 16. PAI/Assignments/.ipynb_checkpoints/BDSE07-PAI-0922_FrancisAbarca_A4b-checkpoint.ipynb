Assignment-04B: Visualization with Matplotlib and Seaborn on Titanic Dataset

Task 1: Load the Dataset and Explore

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('titanic_dataset.csv')
print("First 5 rows of Titanic dataset:")
print(df.head())
```

Task 2: Visualize Passenger Survival

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=df)
plt.title("Count of Survivors (1) and Non-Survivors (0)")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```

Task 3: Analyze Passenger Age Distribution

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Age'].dropna(), bins=30, edgecolor='black')
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Task 4: Explore Class-wise Distribution

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Pclass', data=df)
plt.title("Number of Passengers in Each Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

Task 5: Investigate Fare Distribution

```python
plt.figure(figsize=(8, 5))
sns.boxplot(y='Fare', data=df)
plt.title("Distribution of Passenger Fares")
plt.ylabel("Fare")
plt.show()
```

Task 6: Correlation Between Variables

```python
plt.figure(figsize=(10, 8))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix of Numerical Variables")
plt.show()
```

Task 7: Visualize Passenger Class and Age

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['Age'], df['Pclass'], alpha=0.6)
plt.title("Passenger Age vs Class")
plt.xlabel("Age")
plt.ylabel("Passenger Class")
plt.gca().invert_yaxis()  # so class 1 is at the top
plt.show()
```
