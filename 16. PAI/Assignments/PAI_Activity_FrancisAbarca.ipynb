Task 1: NumPy Arrays

```
import numpy as np

# Creating a 1D array
array_1d = np.array([10, 20, 30, 40, 50])
print("1D Array:\n", array_1d)
print("Dimension:", array_1d.ndim)
print("Shape:", array_1d.shape)
print("Size:", array_1d.size)
print()

# Creating a 2D array
array_2d = np.array([[1, 2], [3, 4], [5, 6]])
print("2D Array:\n", array_2d)
print("Dimension:", array_2d.ndim)
print("Shape:", array_2d.shape)
print("Size:", array_2d.size)
print()

# Creating a 3D array with shape
array_3d = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])
print("3D Array:\n", array_3d)
print("Dimension:", array_3d.ndim)
print("Shape:", array_3d.shape)
print("Size:", array_3d.size)
print()
```

Task 2: Tips.csv Dataset

```
import pandas as pd
import matplotlib.pyplot as plt

# Dataset loading
df = pd.read_csv("tips.csv")

# Print the first 5 rows
print("First 5 rows of the dataset:")
print(df.head())

# Scatter plot
plt.figure(figsize=(8,5))
plt.scatter(df["total_bill"], df["tip"], color='blue', alpha=0.6)
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Scatter Plot of Total Bill vs Tip")
plt.grid(True)
plt.show()
```
